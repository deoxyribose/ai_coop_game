```python
import jax.numpy as jnp
import jax.random as random
import matplotlib.pyplot as plt
```

```python
# sample three points in the plane

foo = random.uniform(random.PRNGKey(0), (10, 3, 2))
print(foo[0])
# sort foo by the first element of the last dimension
idxs = jnp.argsort(foo[..., 0], axis=-1)
foo = foo[jnp.arange(10)[:, None], idxs]
```

```python
dets = jnp.linalg.det(jnp.concat([foo, jnp.ones((10, 3, 1))], axis=-1))
```

```python
for i in range(10):
    plt.figure()
    plt.scatter(*foo[i].T)
    if dets[i] > 0:
        plt.title(f"points downn")
    else:
        plt.title(f"points up")
```
